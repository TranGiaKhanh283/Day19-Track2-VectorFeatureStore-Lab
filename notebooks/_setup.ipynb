```python
"""Path bootstrap for lab notebooks.

Resolves the repo root (where `app/`, `scripts/`, `data/` live) regardless of
where Jupyter was launched from. Used by all 4 notebooks:

    import _setup  # noqa: F401   -- adds repo root to sys.path

When imported from `_setup.py`, `__file__` pins the path. When you open this
as `_setup.ipynb`, Jupyter often has no `__file__` — the code walks up from
the notebook server's cwd until it finds `app/` + `data/` (or `scripts/`).
"""
from __future__ import annotations
```

```python
import sys
from pathlib import Path
import _setup
```

```python
def _find_repo_root() -> Path:
    try:
        here = Path(__file__).resolve()
        cand = here.parent.parent
        if (cand / "app").is_dir():
            return cand
    except NameError:
        pass
    start = Path.cwd().resolve()
    for p in (start, *start.parents):
        if (p / "app").is_dir() and ((p / "data").is_dir() or (p / "scripts").is_dir()):
            return p
    return start


_REPO_ROOT = _find_repo_root()
if str(_REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(_REPO_ROOT))

print(f"_setup: repo root = {_REPO_ROOT}")
```
